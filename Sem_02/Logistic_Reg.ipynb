### In Synthetic data

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
````

````python
X, y = make_classification(n_samples =  100, n_features = 2 , n_redundant = 0, n_informative = 2, n_clusters_per_class = 1, n_classes = 2, random_state = 1)

print(X.shape, y.shape)
_, ax = plt.subplots(figsize=( 6 , 5 )) ##modificación de ejes para definir tamaños
ax.scatter( X[:,0] , X[:,1] , c = y , cmap = 'Paired') ##Scater = tipo de grafico a generar (diagrama de disporsión) ##c: depende de labels de y
````

````python
# Create an instance
logistic_regression = LogisticRegression()

# Train the model
logistic_regression.fit ( X , y ) #(# Samples, # Features) ("100 muestras. 2 carateristicas")
````

````python
predictions = logistic_regression.predict(X)
print(predictions.shape)
````

### **Decision boundaries**

````python
# Min-Max values for the grid of decision boundaries
h_min, h_max = X[:,0].min() , X[:,0].max()
v_min, v_max = X[:,1].min() , X[:,1].max()

# Create grid
h_grid, v_grid = np.meshgrid( np.linspace(h_min,h_max,300) , np.linspace(v_min, v_max,300)) #construye cuadricula para el plano entre los valores minimos del grupo de datos
print(h_grid.shape, v_grid.shape)
print(h_grid.ravel().shape, v_grid.ravel().shape)
print(np.c_[h_grid.ravel(), v_grid.ravel()].shape)
# Estimate predictions over the grid
predictions = logistic_regression.predict(np.c_[h_grid.ravel(), v_grid.ravel()])
print (predictions.shape)

predictions = predictions.reshape(h_grid.shape)
print(predictions.shape)
````

````python

# Draw grid
_, ax = plt.subplots (figsize=(6,5))
ax.pcolormesh( h_grid , v_grid , predictions , cmap = "Paired")

# Scatter real values
ax.scatter ( X[:, 0] , X[:, 1] , c = y , edgecolors = 'k', cmap = "Paired") ##edgecolor = borde de puntos del diagrama

ax.set_xticks(())
ax.set_yticks(())
````
